```python
def maximum_shearing_correlation_distance(dataframe, column1, column2):
    series1 = dataframe[column1].values
    series2 = dataframe[column2].values

    
    # Inicialization
    max_distance = np.inf
    max_shift = 0
    
    max_date = None
    
    # Calcular la distancia para cada desplazamiento
    for shift, date in enumerate(dataframe['date']):
        distance = np.sum(np.abs(series1 - np.roll(series2, shift)))
        if distance < max_distance:
            max_distance = distance
            max_shift = shift
            max_date = date
    
    return max_distance, max_shift, max_date
```

```python
# Create a list with all possible combinations of columns
column_combinations = list(itertools.combinations(df_normalized.columns[1:], 2))  # Exclude the 'date' column

# Calculate the maximum shearing correlation distance for each column combination
for column_pair in column_combinations:
    max_dist, shift, max_date = maximum_shearing_correlation_distance(df_normalized, column_pair[0], column_pair[1])
    print("Maximum shearing correlation distance between '{}' and '{}':".format(column_pair[0], column_pair[1]), max_dist)
    print("Shifting:", shift)
    print("Date:", max_date)
    print()
    
print("We stay with the time series of open and volume")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

# Generamos datos de ejemplo (reemplzalos con tus propios datos)
# Asumiendo que df_normalized es el DataFrame original con las columnas 'date' y 'close'
df_nuevo = df_normalized.copy()

size = int(len(df_nuevo) * 0.8)
plt.figure(figsize=(10, 10))
plt.plot(df_nuevo['date'], df_nuevo['close'])
plt.xlabel('Fecha')
plt.ylabel('Valor de Cierre')
plt.title('Serie Temporal')
plt.show()
print(len(df_nuevo))
# Dividimos los datos en entrenamiento y prueba
train_data = df_nuevo[:size]['close']
test_data = df_nuevo[size:len(df_nuevo)]['close']



# Graficamos la funcin de autocorrelacin (ACF) y la funcin de autocorrelacin parcial (PACF)
plt.figure(figsize=(10, 10))
plot_acf(train_data)
plt.title('Funcin de Autocorrelacin (ACF)')
plt.show()

plt.figure(figsize=(10, 10))
plot_pacf(train_data)
plt.title('Funcin de Autocorrelacin Parcial (PACF)')
plt.show()

# Creamos el modelo ARIMA
order = (1, 1,0)  # (p, d, q)
model = ARIMA(train_data, order=order)

# Ajustamos el modelo ARIMA a los datos de entrenamiento
fit_model = model.fit()

# Hacemos predicciones
forecastTrain = fit_model.forecast(steps=len(train_data))

# Hacemos predicciones
forecast= fit_model.forecast(steps=len(test_data))
# Plot residual errors
residuals = pd.DataFrame(fit_model.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()

# Actual vs Fitted
plt.figure(figsize=(10, 10))
plt.plot(df_nuevo['date'], df_nuevo['close'], label='Datos de Entrenamiento')
plt.plot(df_nuevo[:size]['date'], train_data, label='Datos de Prueba')
plt.plot(df_nuevo[:size]['date'], forecastTrain, label='Predicciones', color='red')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Predicciones ARIMA')
plt.legend()
plt.show()
# Visualizamos las predicciones junto con los datos de prueba
plt.figure(figsize=(10, 10))
plt.plot(df_nuevo['date'], df_nuevo['close'], label='Datos de Entrenamiento')
plt.plot(df_nuevo[size:len(df_nuevo)]['date'], test_data, label='Datos de Prueba')
plt.plot(df_nuevo[size:len(df_nuevo)]['date'], forecast, label='Predicciones', color='red')

plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Predicciones ARIMA')
plt.legend()
plt.show()
```

```python
def arima_forecast(history, order):
    # Fit the model
    model = ARIMA(history, order=order)
    model_fit = model.fit()

    output = model_fit.forecast()
    yhat = output[0]
    return yhat



import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import itertools

def autoarima_forecast(history, d_range=(0, 2), p_range=(0, 5), q_range=(0, 5)):
    best_aic = np.inf
    best_order = None

    for d in range(d_range[0], d_range[1] + 1):
        for p in range(p_range[0], p_range[1] + 1):
            for q in range(q_range[0], q_range[1] + 1):
                order = (p, d, q)
                try:
                    model = ARIMA(history, order=order)
                    model_fit = model.fit()
                    aic = model_fit.aic
                    if aic < best_aic:
                        best_aic = aic
                        best_order = order
                except:
                    continue

    if best_order is None:
        raise ValueError("No se pudo encontrar un modelo ARIMA vlido")

    print("Mejor modelo ARIMA:", best_order, " AIC:", best_aic)

    # Ajusta el mejor modelo ARIMA y realiza la prediccin
    model = ARIMA(history, order=best_order)
    model_fit = model.fit()
    forecast = model_fit.forecast()
    yhat = forecast[0]

    return yhat, best_order


```

```python
X = df_normalized.copy()

diferencias = np.diff(X['close'])
size = int(len(diferencias) * 0.8)

df_nuevo = pd.DataFrame({'close': diferencias})

X = X.iloc[1:]

# Concatenar 'date' de X y 'close_diff' de df_nuevo a lo largo de las columnas
X = pd.concat([X['date'].reset_index(drop=True), df_nuevo.reset_index(drop=True)], axis=1)


# Crear una figura y ejes
plt.figure(figsize=(10, 6))
# Graficar la serie temporal diferenciada
plt.plot(X['date'], diferencias, label='Serie Temporal Diferenciada', color='blue')

# Etiquetas y ttulo
plt.xlabel('Fecha')
plt.ylabel('Diferencias')
plt.title('Serie Temporal Diferenciada')
plt.legend()

# Mostrar la grfica
plt.show()


train, test = X['close'][:size], X['close'][size:]
history = [x for x in train]
predictions = list()
best_orders = list()

plt.figure(figsize=(10, 10))
plot_acf(train)
plt.title('Funcin de Autocorrelacin (ACF)')
plt.show()

plt.figure(figsize=(10, 10))
plot_pacf(train)
plt.title('Funcin de Autocorrelacin Parcial (PACF)')
plt.show()

for t in range(len(test)):
    # Generate a prediction
    yhat, best_order = autoarima_forecast(history)
    predictions.append(yhat)
    best_orders.append(best_order)
    # Add the predicted value to the training set
    obs = test.iloc[t]
    history.append(obs)

fig, ax = plt.subplots(2, 1, figsize=(15, 15))  # Crear una figura con dos subgrficos

ax[0].plot(X['date'].iloc[:size], train, label='Datos de Entrenamiento')
ax[0].plot(X['date'].iloc[size:], test, label='Datos de Prueba')

ax[0].set_xlabel('Fecha')
ax[0].set_ylabel('Valor')
ax[0].set_title('Datos de Entrenamiento y Prueba')
ax[0].legend()


ax[1].plot(X['date'][:size], train, label='Datos de Entrenamiento')
ax[1].plot(X['date'][size:], test, label='Datos de Prueba')
ax[1].plot(X['date'][size:], predictions, label='Predicciones', color='red')

ax[1].set_xlabel('Fecha')
ax[1].set_ylabel('Valor')
ax[1].set_title('Predicciones ARIMA')
ax[1].legend()


plt.show()

```
