```python
def maximum_shearing_correlation_distance(dataframe, column1, column2):
    series1 = dataframe[column1].values
    series2 = dataframe[column2].values

    
    # Inicialization
    max_distance = np.inf
    max_shift = 0
    
    max_date = None
    
    # Calcular la distancia para cada desplazamiento
    for shift, date in enumerate(dataframe['date']):
        distance = np.sum(np.abs(series1 - np.roll(series2, shift)))
        if distance < max_distance:
            max_distance = distance
            max_shift = shift
            max_date = date
    
    return max_distance, max_shift, max_date
```

```python
# Create a list with all possible combinations of columns
column_combinations = list(itertools.combinations(df_normalized.columns[1:], 2))  # Exclude the 'date' column

# Calculate the maximum shearing correlation distance for each column combination
for column_pair in column_combinations:
    max_dist, shift, max_date = maximum_shearing_correlation_distance(df_normalized, column_pair[0], column_pair[1])
    print("Maximum shearing correlation distance between '{}' and '{}':".format(column_pair[0], column_pair[1]), max_dist)
    print("Shifting:", shift)
    print("Date:", max_date)
    print()
    
print("We stay with the time series of open and volume")
```

```python
# Definir una lista de patrones de pasos para probar
steps = list(range(1, 7))  # Valores del 1 al 6
types = ['a', 'b', 'c', 'd']  # Valores de la 'a' a la 'd'

# Generar todas las combinaciones posibles de valores de pasos y tipos
step_patterns = list(product(steps, types))

# Inicializar la distancia mínima y el patrón de pasos correspondiente
min_distance = float('inf')
best_pattern = None

# Iterar sobre los patrones de pasos y calcular la distancia DTW para cada uno
for pattern in step_patterns:
    alignment = dtw(series2, series1, keep_internals=True, step_pattern=pattern)
    distance = alignment.distance
    print(f"Distancia DTW con patrón de pasos {pattern}: {distance}")
    
    # Actualizar la distancia mínima y el mejor patrón de pasos si es necesario
    if distance < min_distance:
        min_distance = distance
        best_pattern = pattern

# Trazar la alineación entre las dos series utilizando el mejor patrón de pasos encontrado
alignment = dtw(series2, series1, keep_internals=True, step_pattern=best_pattern)
ax = alignment.plot(type="twoway", offset=-14)
ax.figure.set_size_inches(30, 20)

# Imprimir la distancia mínima y el mejor patrón de pasos
print(f"Distancia mínima DTW: {min_distance} con patrón de pasos {best_pattern}")
```
